```python
a = pd.DataFrame({"id" : [1,2,3,4,5],
                  'Name':['Alax', 'Amy', 'Allen', 'Alice', 'Ava'],
                  'Sub_id': [1,2,4,5,6]})
b = pd.DataFrame({"id" : [1,2,3,4,5],
                  'Name':['Bob', 'Ben', 'Bran', 'Bryan', 'Bri'],
                  'Sub_id': [2,4,3,6,5]})
```

```python
a
```

```python
b
```

```python
pd.merge(a,b, on = 'id')   #inner join
```

```python
pd.merge(a,b, on ='Name')  #inner join
```

```python
pd.merge(a,b, on ='Sub_id')  #inner join
```

```python
pd.merge(a,b, on ='Name', how= 'outer')  #outer join
```

```python
pd.merge(a,b, on ='id', how= 'outer')  #outer join
```

```python
pd.merge(a,b, on ='Sub_id', how= 'outer')  #outer join
```

```python
pd.merge(a,b, on ='Name', how= 'left')  #Left join
```

```python
pd.merge(a,b, on ='id', how= 'left')  #Left join
```

```python
pd.merge(a,b, on ='Sub_id', how= 'left')  #left join
```

```python
pd.merge(a,b, on ='Sub_id', how= 'right')  #right join
```

```python
pd.merge(a,b, on ='id', how= 'right')  #right join
```

```python
pd.merge(a,b, on ='Name', how= 'right')  #right join
```
