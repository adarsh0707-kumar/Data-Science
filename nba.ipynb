```python
def ConvertHeight(data):
  height = data.split("-")
  newHeightF = int (height[0]) * 30.48
  newHeightI = int(height[1]) * 2.54
  heightCm = newHeightF + newHeightI
  return heightCm

df['HeightCm'] = df.Height.apply(ConvertHeight)
```
