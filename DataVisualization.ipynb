Data Visualization 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
t = sns.load_dataset('titanic')
t
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Sine Wave', color='blue')
ax.plot(x,z, label='Cosine Wave', color='red')

ax.legend()

ax.set_xlabel("Angle")
ax.set_ylabel("Co/sine's Angle")
ax.set_title("Co/sine Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = 1/(np.sin(x))
z = 1/(np.cos(x))
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Cosecant Wave', color='blue')
ax.plot(x,z, label='Secant Wave', color='red')  
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Co/secant's Angle")
ax.set_title("Co/secant Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.tan(x)
z = 1/(np.tan(x))
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Tangent Wave', color='blue')
ax.plot(x,z, label='Cotangent Wave', color='red')
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Co/tangent's Angle")
ax.set_title("Co/tangent Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)

y = np.log10(x)
z = np.log2(x)
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Logarithmic Wave of log10', color='blue')
ax.plot(x,z, label='Logarithmic Wave of log2', color='red')
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Logarithmic of Angle")
ax.set_title("Logarithmic Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.tan(x)

fig = plt.figure(figsize= (8,5))
ax1 = fig.add_axes([0,0,1,0.45])
ax2 = fig.add_axes([0,0.5,1,0.55])

ax1.plot(x,y, label='Sine Wave', color='blue')
ax1.legend()
ax1.set_xlabel("Angle")
ax1.set_ylabel("Sine of Angle")
ax1.set_title("Sine Function")


ax2.plot(x,z, label='tan Wave', color='red')
ax2.legend()
ax2.set_ylabel("tan of Angle")
ax2.set_title("tan Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1) 
y = np.sin(x)
z = np.cos(x)
p = np.tan(x)
q = np.log10(x)

fig = plt.figure(figsize=(10,7))

ax1 = fig.add_axes([0,0,0.45,0.45])
ax2 = fig.add_axes([0.55,0,0.45,0.45])
ax3 = fig.add_axes([0,0.55,0.45,0.45])
ax4 = fig.add_axes([0.55,0.55, 0.45,0.45])

ax1.plot(x,y, label='Sine', color='green')
ax2.plot(x,z, label='Cosine', color= 'black')
ax3.plot(x,p, label='Tan', color= 'red')
ax4.plot(x,q, label='Log')

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()
```

```python
ta = t.groupby('sibsp').age.mean()
ta
```

```python
x = ta.index
y = ta.values

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(x,y)

for i, value in zip(x,y):
  plt.text(i, value - 4, str(round(value, 2)), ha='center', color = 'yellow', fontweight = 800, fontsize=12)
```

```python
tp = t.pclass.value_counts()
tp
```

```python
x = tp.index
y = tp.values

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.set_xticks([1,2,3], ['1st Class', '2nd Class', '3rd Class'])
ax.bar(x,y)

for i, value in zip(x,y):
  plt.text(i, value -50, str(value), ha='center', color = 'yellow', fontweight = 800, fontsize=12)
```

```python
tsibsp = t.groupby('sibsp').age.mean()

tsex = t.groupby('sex').age.mean()

tsurvived = t.groupby('survived').age.mean()

tparch = t.groupby('parch').age.mean()

fig = plt.figure(figsize=(10,7))


ax1 = fig.add_axes([0,0,0.45,0.45])
ax2 = fig.add_axes([0.55,0,0.45,0.45])
ax3 = fig.add_axes([0,0.55,0.45,0.45])
ax4 = fig.add_axes([0.55,0.55, 0.45,0.45])


ax1.bar(tsibsp.index, tsibsp.values , label='Sibsp')
ax2.bar(tsex.index, tsex.values, width=0.5, label='Sex')
ax3.bar(tsurvived.index, tsurvived.values, width=0.5, label='Survived')
ax4.bar(tparch.index, tparch.values, label='Parch')

ax3.set_xticks([0,1], ['Survived', 'Not Survived'])

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()



plt.show()
```
