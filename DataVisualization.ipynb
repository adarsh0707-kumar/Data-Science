```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Sine Wave', color='blue')
ax.plot(x,z, label='Cosine Wave', color='red')

ax.legend()

ax.set_xlabel("Angle")
ax.set_ylabel("Co/sine's Angle")
ax.set_title("Co/sine Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = 1/(np.sin(x))
z = 1/(np.cos(x))
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Cosecant Wave', color='blue')
ax.plot(x,z, label='Secant Wave', color='red')  
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Co/secant's Angle")
ax.set_title("Co/secant Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.tan(x)
z = 1/(np.tan(x))
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Tangent Wave', color='blue')
ax.plot(x,z, label='Cotangent Wave', color='red')
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Co/tangent's Angle")
ax.set_title("Co/tangent Function")
plt.show()
```

```python
x = np.arange(0, 2*np.pi, 0.1)

y = np.log10(x)
z = np.log2(x)
fig = plt.figure(figsize= (8,4))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Logarithmic Wave of log10', color='blue')
ax.plot(x,z, label='Logarithmic Wave of log2', color='red')
ax.legend()
ax.set_xlabel("Angle")
ax.set_ylabel("Logarithmic of Angle")
ax.set_title("Logarithmic Function")
plt.show()
```

Write a program to rotate a list n time without using extra space

```python
def reverse(arr, start, end):
  while start < end:
    arr[start], arr[end] = arr[end], arr[start]
    start += 1
    end -= 1


def rotate(arr, n):
  size = len(arr)
  n = n% size #in case n> size
  # reverse whole arr
  reverse(arr,0,size-1)
  
  #reverse the first n element 

  reverse(arr,0, n-1)

  reverse(arr, n, size-1)

arr =[1,2,3,4,5,6,7]
n = 3
rotate(arr,n)
print("Rotated List", arr)
```

Find the 2nd largest unique element from a list without using max() or short()

```python
def second_largest_unique(arr):
  largest = float('-inf')
  second = float('-inf')
  for num in arr:
    if num > largest:
      second = largest
      largest = num
    elif num > second and num != largest:
      second = num
  if  second == float('-inf'):
    return None
  return second

arr = [10,20,4,20,10,5]
print("Second largest unique element: ", second_largest_unique(arr))
```

```python
def second_largest_unique(lst):
    unique_values = set(lst)
    if len(unique_values) < 2:
        return None  

    largest = second = float('-inf')

    for num in lst:
        if num > largest:
            second, largest = largest, num
        elif num != largest and num > second:
            second = num

    return second

arr = [10,20,4,20,10,5]
print("Second largest unique element: ", second_largest_unique(arr))
```

Write a dic ptogram that group the words using anagram

```python
def group_anagrams(words):
  anagram_map = {}
  for word in words:
    key = ''.join(sorted(word))  
    if key in anagram_map:
      anagram_map[key].append(word)
    else:
      anagram_map[key] = [word]
  return list(anagram_map.values())


words = ["eat", "tea", "tan", "ate", "nat", "bat"]
print(group_anagrams(words))
```
