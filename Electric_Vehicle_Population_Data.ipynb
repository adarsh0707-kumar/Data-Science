```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Electric_Vehicle_Population_Data.csv")
df
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.countplot(
    x="Model Year",
    hue="Electric Vehicle Type",
    data=df,
    ax=ax
)

# Annotate each bar manually
for patch in ax.patches:
    height = patch.get_height()
    ax.text(
        patch.get_x() + patch.get_width() / 2,  # x coordinate: center of bar
        height - 0.2,                           # y coordinate: just above bar
        int(height),                            # label text
        ha='center'                             # horizontal alignment
    )

plt.title("Count of EV Types by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.legend(title="EV Type")
plt.tight_layout()
plt.show()
```

```python
counts = df["Model Year"].value_counts()
plt.figure(figsize=(10, 10))
plt.bar(counts.index.astype(str), counts.values)
plt.title("Count of Entries by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
